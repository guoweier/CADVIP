## Arrange Populus Old.Name and Seq.Name 

The dataset corresponded with Populus Old.Name and Seq.Name has Seq.Name in order and corresponds to Old.Name (field name). In order to add seq.name to bulk samples for K-mer analysis, we have to get a list of unique Old.Name which corresponds to one or more Seq.Names.  

This Notebook is aimed to transfer the original format into target format. Later, this can be used to label the bulk samples. (Prepare for re-sequencing)

```python
#use iPython magin notation to make the plots appear inline
%matplotlib inline
#import all needed libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
oldseq_or = open("../output/removed_rep_20220608.csv")

head = oldseq_or.readline()
all_dict = {}

for line in oldseq_or:
    line = line.split("\n")[0]
    ln = line.split(",")
    if ln[0] not in all_dict:
        all_dict[ln[0]] = []
    all_dict[ln[0]] += [ln[1]]
    
oldseq_or.close()
```

```python
all_dict
```

```python
o = open("../output/old_name_unique.txt", "w")
header = "Old.Name"+"\t"+"Seq.Name"
o.write(header+"\n")

for key in all_dict:
    seqlibs = ""
    for value in all_dict[key]:
        seqlibs += value
    text = key+"\t"+seqlibs
    o.write(text+"\n")
o.close()
```
